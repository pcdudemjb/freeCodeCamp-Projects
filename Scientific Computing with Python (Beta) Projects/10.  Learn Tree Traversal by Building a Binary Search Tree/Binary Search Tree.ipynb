```python
class TreeNode:
    def __init__(self, key):
        # Initialize a tree node with a key, left, and right pointers
        self.key = key
        self.left = None
        self.right = None

    def __str__(self):
        # Convert the node to a string representation
        return str(self.key)


class BinarySearchTree:
    def __init__(self):
        # Initialize an empty binary search tree with no root
        self.root = None

    def insert(self,key):
        # Insert a key into the binary search tree
        self.root = self._insert(self.root, key)

    def _insert(self, node, key):
        # Helper function for inserting a key recursively
        if node is None:
            # If the current node is None, create a new node with the given key
            return TreeNode(key)
        if key < node.key:
            # If the key is less than the current node's key, go to the left subtree
            node.left = self._insert(node.left, key)
        elif key > node.key:
            # If the key is greater than the current node's key, go to the right subtree
            node.right = self._insert(node.right, key)
        return node

    def search(self, key):
        # Search for a key in the binary search tree
        return self._search(self.root, key)

    def _search(self, node, key):
        if node is None or node.key == key:
            # If the current node is None or has the target key, return the node
            return node
        if key < node.key:
            # If the key is less than the current node's key, search in the left subtree
            return self._search(node.left, key)
        return self._search(node.right, key)

    def delete(self, key):
        # Deletes a key from the binary search tree
        self.root = self._delete(self.root, key)

    def _delete(self, node, key):
        # Helper function for deleting a key recursively
        if node is None:
            # If the current node is None, return None
            return node
        if key < node.key:
            # If the key is less than the current node's key, delete from the left subtree
            node.left = self._delete(node.left, key)
        elif key > node.key:
            # If the key is greater than the current node's key, delete from the right subtree
            node.right = self._delete(node.right, key)
        else:
            if node.left is None:
                # If the left child is None, replace the node with its right child
                return node.right
            elif node.right is None:
                # If the right child is None, replace the node with its left child
                return node.left
            # If both children are present, replace the node with the minimum value from the right subtree
            node.key = self._min_value(node.right)
            node.right = self._delete(node.right, node.key)
        return node

    def _min_value(self, node):
        # Finds the minimum value in a subtree
        while node.left is not None:
            # Traverse to the leftmost node to find the minimum value
            node = node.left
        return node.key

    def inorder_traversal(self):
        # Performs an inorder traversal of the binary search tree
        result = []
        self._inorder_traversal(self.root, result)
        return result

    def _inorder_traversal(self, node, result):
        # Helper function for inorder traversal
        if node:
            # Recursively traverse the left subtree, visit the node, and traverse the right subtree
            self._inorder_traversal(node.left, result)
            result.append(node.key)
            self._inorder_traversal(node.right, result)

            
# Example usage
bst = BinarySearchTree()
nodes = [50, 30, 20, 40, 70, 60, 80]

for node in nodes:
    # Insert nodes into the binary search tree
    bst.insert(node)
print("Inorder traversal:", bst.inorder_traversal())
print("Search for 40:", bst.search(40))
# Delete a node and print the inorder traversal again.
bst.delete(40)
print("Inorder traversal after deleting 40:", bst.inorder_traversal())
print("Search for 40:", bst.search(40))
```
